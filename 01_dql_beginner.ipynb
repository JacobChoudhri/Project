### WHERE & Comparison Operators

The WHERE clause in SQL allows you to add a specific criteria that column values must meet in order to be in the query result.

```python
%%sql
SELECT *
FROM people_df
WHERE region = 'West'
```

In the query above, I queried all of the rows from the people_df dataframe where the region was "West". The * allows for me to get all of the columns in that table where the region is "West". If I changed the * to the name of the column 'Region', I would only be querying the one column where "West" is.


Like so:

```python
%%sql
SELECT region
FROM people_df
WHERE region = 'West'
```

You probably noticed the usage of the equal sign, which in SQL is called a comparison operator. There are many other comparison operators that you may recognize, such as <>, <, >, <=, >=. All of which are used to make comparisons in your query.

```python
%%sql
SELECT sales
FROM orders_df
WHERE sales > 500
LIMIT 10
```

In the above example I queried all of the sales that exceeded 500. These compairison operators can be very useful when trying to find a specific section of data without it being lost in all of the other results.

#### Get all of the records from the Orders table where the quantity is greater than or equal to 10. Limit your query to 5 results.

```python
%%sql
SELECT *
FROM orders_df
WHERE quantity >= 10
LIMIT 5

```

### Logical Operators

Logical operators in SQL are used to combine conditions in a WHERE clause to filter data based on multiple conditions. There are many useful logical operators in SQL. For right now we will focus on three of the most used ones: AND, OR, and LIKE. Others that you may see if you continue with your SQL journey are IN, BETWEEN, and NOT.

* The AND operator is binding on both conditions being true.
* The OR operator is not binding and only requires one of the conditions to be true.
* The LIKE operator in SQL is used for pattern matching in a WHERE clause. It allows you to search for a specified pattern of characters within a column.
    


##### AND

```python
%%sql
SELECT *
FROM orders_df
WHERE region = 'West' AND category = 'Technology'
LIMIT 5
```

The results show all of the rows where the order was made in the western region AND where the category of the item purchased was "Technology".

##### OR

```python
%%sql
SELECT *
FROM orders_df
WHERE sub_category = 'Paper' OR state = 'New York'
LIMIT 5
```

The results show all of the rows where the order was under the sub-category was 'Paper' OR where the state was New York.

##### LIKE

The LIKE operator is vital in SQL because it allows for pattern matching within string values, providing flexible and powerful search capabilities. It enables SQL queries to retrieve data based on partial or similar matches, making it invaluable for tasks such as filtering, searching, and sorting textual data efficiently and accurately. Without the LIKE operator, it would be challenging to perform complex string comparisons and locate specific data patterns within a database.

```python
%%sql
SELECT *
FROM orders_df
WHERE customer_id LIKE 'ME%'
LIMIT 10
```

As you can see this can be a very powerful tool especially when trying to character match on a massive dataset much bigger than the one we're using. In the above example I simply queried all results where the customer_id starts with the characters 'ME'. 
Below I've added some context as to how the % operator works.

* WHERE CustomerName LIKE 'a%'   -->	  Finds any values that start with "a"
* WHERE CustomerName LIKE '%a'   -- >	  Finds any values that end with "a"
* WHERE CustomerName LIKE '%or%'  -->	  Finds any values that have "or" in any position
* WHERE CustomerName LIKE '_r%'  -- >	  Finds any values that have "r" in the second position
* WHERE CustomerName LIKE 'a_%'  -- >	  Finds any values that start with "a" and are at least 2 characters in length
* WHERE CustomerName LIKE 'a__%'  -->	  Finds any values that start with "a" and are at least 3 characters in length
* WHERE ContactName LIKE 'a%o'  --	 >   Finds any values that start with "a" and ends with "o"

### ORDER BY

ORDER BY is a clause that allows you to order your queried results in ascending or descending order. It enables the arrangement of data in a meaningful and organized manner, making it easier to analyze and interpret the results. The ORDER BY clause is crucial for presenting data in a logical sequence and facilitating efficient decision-making processes when working with large datasets. Its important to keep in mind that the ORDER BY clause should come after the WHERE clause.

```python
%%sql
SELECT *
FROM people_df
ORDER BY region ASC
```

```python
%%sql
SELECT *
FROM people_df
ORDER BY region DESC
```

Obviously those were very simple examples but you can see why ordering your results may be helpful. If the result set contained 100 results, it would be nice to know where to be looking if your only looking for one record in particular.

### Aggregate Functions

In SQL, aggregate functions are powerful tools used to perform calculations on sets of data and return a single result. They enable us to derive meaningful insights from large datasets by summarizing or grouping data. Two essential components in aggregate function queries are the GROUP BY clause and the HAVING clause. The GROUP BY clause is used to group rows based on one or more columns, creating distinct groups. On the other hand, the HAVING clause allows us to filter the grouped data based on specific conditions, acting as a conditional filter for the groups.

##### GROUP BY

```python
%%sql
SELECT customer_id, SUM(quantity)
FROM orders_df
GROUP BY customer_id
ORDER BY SUM(quantity) DESC
LIMIT 10


```

I think this is a good example to show how much you can do with these simple commands. In this example I selected each customer by their id, as well as their quantity for their order. However in this example I was able to use the SUM function around the quantity column to get the total amount of ALL orders. Using GROUP BY it was able to place each total with the respective customer id. I then made sure to order it in a descending manner in order to see who has the most orders. Just 5 lines of code can help you find so many small pieces of information that together, show a whole picture.

#### HAVING

Continuing from the previous example, let's say we only want to retrieve the customers who have ordered a total quantity LESS than 20.

```python
%%sql
SELECT customer_id, SUM(quantity)
FROM orders_df
GROUP BY customer_id
HAVING SUM(quantity) < 10
LIMIT 10
```

### Data types

In SQL, data types define the type of data that can be stored in a column or variable. They determine the range of values, the storage requirements, and the operations that can be performed on the data. Here are some common data types found in SQL:

* INTEGER: Used to store whole numbers (positive or negative) without decimal places.
* FLOAT/REAL: Used to store floating-point numbers with decimal places.
* VARCHAR/CHAR: Used to store variable-length or fixed-length character strings, respectively.
* DATE/DATETIME: Used to store dates or a combination of dates and times.
* BOOLEAN: Used to store true/false or binary values.


In SQL, the CAST function is used to convert data from one data type to another. It allows you to explicitly specify the target data type for a given value or expression. For example, you can use CAST to convert a string into an integer or a date into a string. Here's an example:




```python
%%sql
SELECT order_id, customer_id, CAST(order_date AS DATE) AS ConvertedDate
FROM orders_df
LIMIT 5
```

I was able to convert the date column into a easier to read way for SQL using the DATE function. I then renamed it "ConvertedDate" using the AS clause.
